```python
'''
~~~~~~~~~~~~PP Mini Project~~~~~~~~~~~~~~~

        Made By:
        658 Avdhoot Tavhare (F3 Batch - 0120200105)
        659 Rohit Pingale (F3 Batch - 0120200159)

~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
'''


import tkinter as tk   #importing TkInter library from python , and abbreviating as tk 
from tkinter import *   #from TkInter Library, we import all the sub library 
import random    #this library is used to generate random numbers in the "bill number" text field, every time the user will execute the code
                #new number will be generated ranging from 0 to 9999

class Grocery_Bill_App:
    def __init__(self,window):
        self.window=window 
        self.window.geometry("1300x700")  #.geometry decides the geometry or the layout of the window of the GUI for the user to interact with
        self.window.title("Whole Super Market")  #title for the main GUI window 
        
       #~~~~~~~~~~~~~~~~~~ Intializing variables required in labels, buttons etc.~~~~~~~~~~~~~~~
        self.name = StringVar()
        self.phno = IntVar()
        #For Generating Random Bill Numbers
        ran=random.randint(0,9999)
        self.BillNo=StringVar()
        
        self.BillNo.set(str(ran)) #setting the BillNo value to a random variable so that whenever the code is executed, bill no
                                  #text field will generate random values between 0 and 9999

        self.veg1value=IntVar()
        self.veg2value=IntVar()
        self.veg3value=IntVar()
        self.veg4value=IntVar()
        self.veg5value=IntVar()
        self.fr1value=IntVar()
        self.fr2value=IntVar()
        self.fr3value=IntVar()
        self.fr4value=IntVar()
        self.fr5value=IntVar()
        self.d1value=IntVar()
        self.d2value=IntVar()
        self.d3value=IntVar()
        self.d4value=IntVar()
        self.d5value=IntVar()
        self.cosm1value=IntVar()
        self.cosm2value=IntVar()
        self.cosm3value=IntVar()
        self.cosm4value=IntVar()
        self.cosm5value=IntVar()

        self.totalveg=StringVar()
        self.totalfruit=StringVar()
        self.totaldairy=StringVar()
        self.totalcosm=StringVar()
        
        self.genbill=StringVar()

        self.canvas=tk.Canvas(self.window, height=700, width=1300, bg="#DAF7A6")
        self.canvas.pack()   #.pack() geometry manager organizes widgets in blocks before placing them in the parent widget.

        self.frame=tk.Frame(self.window, bg="#FFFFFF")  #hex code for white color
        self.frame.place(relwidth=0.8, relheight=0.8, relx=0.1, rely=0.1)  #relwidth and relheight will decide the height and width of the frame

        #~~~~~~~~~~~~~This code is for the heading of the Frame~~~~~~~~~~~~~~~~~
        
        shop_name=Label(self.frame, text="Whole Super Market", font=("Calibri", 30, "bold"), fg="#000000")  #hex code color for black
        shop_name.pack()
        
        #~~~~~~~~~~~~~~~~~~~~~~~Label created for information of the Group Members~~~~~~~~~~~~~~~~~~~~
        
        group_info=Label(self.frame, text="Group 5", font=("Calibri", 20, "bold"), fg="#000000")
        group_info.place(x=825, y=5)
        
        grp1_info=Label(self.frame, text="658: Avdhoot Tavhare - 0120200105", font=("Calibri", 12), fg="#000000")
        grp1_info.place(x=780, y=50)
        
        grp2_info=Label(self.frame, text="659: Rohit Pingale - 0120200159", font=("Calibri", 12), fg="#000000")
        grp2_info.place(x=780, y=75)
        
        app_info=Label(self.frame, text="This Application allows the user to choose the required number of products \n from each section of the Super Market. It will sum up all the\n quantites entered in the Text Fields, and it will generate the final bill into \n the Total Bill Text Area", font=("Calibri",12), fg="#000000")
        app_info.place(x=5,y=160)
        #~~~~~~~~~~~~Customer Name label And Corresponding Text Field~~~~~~~~~~~~~~~~
        
        self.CustName=Label(self.frame, text="Customer Name: ", font=("Calibri", 12))
        self.CustName.place(x=5,y=70)
        self.CustName_input=Entry(self.frame, textvariable=self.name, font=("Calibri",10), width=20, bd=2)
        self.CustName_input.place(x=160, y=70)

        #~~~~~~~~~~~~~Customer Phone and corresponding Text field~~~~~~~~~~~~~~~
        
        self.phno=Label(self.frame, text="Phone Number: ", font=("Calibri", 12))
        self.phno.place(x=5,y=100)
        self.PhNo_input=Entry(self.frame, textvariable=self.phno, font=("Calibri", 10), width=20, bd=2)
        self.PhNo_input.place(x=160,y=100)

        #~~~~~~~~~~~~~~~~~~Customer Bill No and corresponding Text field~~~~~~~~~~~~~~~~~~
        
        self.billno=Label(self.frame, text="Bill Number: ", font=("Calibri", 12))
        self.billno.place(x=5,y=130)
        self.BillNo_input=Entry(self.frame, textvariable=self.BillNo, font=("Calibri", 10), width=20, bd=2)
        self.BillNo_input.place(x=160,y=130)
        
        #~~~~~~~~~~~~~~~~~Vegetable Frame~~~~~~~~~~~~~~~
        
        veg_frame=LabelFrame(self.frame, text="Vegetables", bd=3, fg="#000000", font=("Calibri", 12, "bold"))
        veg_frame.place(x=10, y=275, width=175, height=200)

        #~~~~~~~~~~~~~~Frame Content~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
        
        veg1=Label(veg_frame, text="Potato",font=("Calibri", 10))
        veg1.grid(row=0, column=0, padx=5, pady=5)
        veg1_entry=Entry(veg_frame, bd=2, textvariable=self.veg1value, width=7)
        veg1_entry.grid(row=0, column=1, ipadx=3, ipady=3)

        veg2=Label(veg_frame, text="Cabbage",font=("Calibri", 10))
        veg2.grid(row=1, column=0, padx=5, pady=5)
        veg2_entry=Entry(veg_frame, bd=2, textvariable=self.veg2value, width=7)
        veg2_entry.grid(row=1, column=1, ipadx=3, ipady=3)

        veg3=Label(veg_frame, text="Tomato",font=("Calibri", 10))
        veg3.grid(row=2, column=0, padx=5, pady=5)
        veg3_entry=Entry(veg_frame, bd=2, textvariable=self.veg3value, width=7)
        veg3_entry.grid(row=2, column=1, ipadx=3, ipady=3)

        veg4=Label(veg_frame, text="Onions",font=("Calibri", 10))
        veg4.grid(row=3, column=0, padx=5, pady=5)
        veg4_entry=Entry(veg_frame, bd=2, textvariable=self.veg4value, width=7)
        veg4_entry.grid(row=3, column=1, ipadx=3, ipady=3)

        veg5=Label(veg_frame, text="Baby Corn",font=("Calibri", 10))
        veg5.grid(row=4, column=0, padx=5, pady=5)
        veg5_entry=Entry(veg_frame, bd=2, textvariable=self.veg5value, width=7)
        veg5_entry.grid(row=4, column=1, ipadx=3, ipady=3)

        #~~~~~~~~~~~~~~~~~~~Fruits Frame~~~~~~~~~~~~~~~~~~~
        
        fruit_frame=LabelFrame(self.frame, text="Fruits", bd=3, fg="#000000", font=("Calibri", 12, "bold"))
        fruit_frame.place(x=185, y=275, width=175, height=200)

        #~~~~~~~~~~~~~~~~~~Frame Content~~~~~~~~~~~~~~~~~~~~~~~~~~
        
        fr1=Label(fruit_frame, text="Apple",font=("Calibri", 10))
        fr1.grid(row=0, column=0, padx=5, pady=5)
        fr1_entry=Entry(fruit_frame, bd=2, textvariable=self.fr1value, width=7)
        fr1_entry.grid(row=0, column=1, ipadx=3, ipady=3)

        fr2=Label(fruit_frame, text="Pear",font=("Calibri", 10))
        fr2.grid(row=1, column=0, padx=5, pady=5)
        fr2_entry=Entry(fruit_frame, bd=2, textvariable=self.fr2value, width=7)
        fr2_entry.grid(row=1, column=1, ipadx=3, ipady=3)

        fr3=Label(fruit_frame, text="Watermelon",font=("Calibri", 10))
        fr3.grid(row=2, column=0, padx=5, pady=5)
        fr3_entry=Entry(fruit_frame, bd=2, textvariable=self.fr3value, width=7)
        fr3_entry.grid(row=2, column=1, ipadx=3, ipady=3)

        fr4=Label(fruit_frame, text="Pomegranate",font=("Calibri", 10))
        fr4.grid(row=3, column=0, padx=5, pady=5)
        fr4_entry=Entry(fruit_frame, bd=2, textvariable=self.fr4value, width=7)
        fr4_entry.grid(row=3, column=1, ipadx=3, ipady=3)

        fr5=Label(fruit_frame, text="Kiwi",font=("Calibri", 10))
        fr5.grid(row=4, column=0, padx=5, pady=5)
        fr5_entry=Entry(fruit_frame, bd=2, textvariable=self.fr5value, width=7)
        fr5_entry.grid(row=4, column=1, ipadx=3, ipady=3)

        #~~~~~~~~~~~~~~~~Dairy Frame~~~~~~~~~~~~~~~~~~
        
        dairy_frame=LabelFrame(self.frame, text="Dairy", bd=3, fg="#000000", font=("Calibri", 12, "bold"))
        dairy_frame.place(x=360, y=275, width=175, height=200)

        #~~~~~~~~~~~~~~~~~~~~Frame Content~~~~~~~~~~~~~~~~~~~~~~
        
        d1=Label(dairy_frame, text="Milk",font=("Calibri", 10))
        d1.grid(row=0, column=0, padx=5, pady=5)
        d1_entry=Entry(dairy_frame, bd=2, textvariable=self.d1value, width=7)
        d1_entry.grid(row=0, column=1, ipadx=3, ipady=3)

        d2=Label(dairy_frame, text="Yogurt",font=("Calibri", 10))
        d2.grid(row=1, column=0, padx=5, pady=5)
        d2_entry=Entry(dairy_frame, bd=2, textvariable=self.d2value, width=7)
        d2_entry.grid(row=1, column=1, ipadx=3, ipady=3)

        d3=Label(dairy_frame, text="Cheese",font=("Calibri", 10))
        d3.grid(row=2, column=0, padx=5, pady=5)
        d3_entry=Entry(dairy_frame, bd=2, textvariable=self.d3value, width=7)
        d3_entry.grid(row=2, column=1, ipadx=3, ipady=3)

        d4=Label(dairy_frame, text="Butter",font=("Calibri", 10))
        d4.grid(row=3, column=0, padx=5, pady=5)
        d4_entry=Entry(dairy_frame, bd=2, textvariable=self.d4value, width=7)
        d4_entry.grid(row=3, column=1, ipadx=3, ipady=3)

        d5=Label(dairy_frame, text="Custard",font=("Calibri", 10))
        d5.grid(row=4, column=0, padx=5, pady=5)
        d5_entry=Entry(dairy_frame, bd=2, textvariable=self.d5value, width=7)
        d5_entry.grid(row=4, column=1, ipadx=3, ipady=3)
        
        
        #~~~~~~~~~~~~~~~~~~~~~~Cosmetics Frame~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
        
        cosm_frame=LabelFrame(self.frame, text="Cosmetics", bd=3, fg="#000000", font=("Calibri", 12, "bold"))
        cosm_frame.place(x=535, y=275, width=175, height=200)
        
        cosm1=Label(cosm_frame, text="Face Cream",font=("Calibri", 10))
        cosm1.grid(row=0, column=0, padx=5, pady=5)
        cosm1_entry=Entry(cosm_frame, bd=2, textvariable=self.cosm1value, width=7)
        cosm1_entry.grid(row=0, column=1, ipadx=3, ipady=3)

        cosm2=Label(cosm_frame, text="Body Soap",font=("Calibri", 10))
        cosm2.grid(row=1, column=0, padx=5, pady=5)
        cosm2_entry=Entry(cosm_frame, bd=2, textvariable=self.cosm2value, width=7)
        cosm2_entry.grid(row=1, column=1, ipadx=3, ipady=3)

        cosm3=Label(cosm_frame, text="Body Lotion",font=("Calibri", 10))
        cosm3.grid(row=2, column=0, padx=5, pady=5)
        cosm3_entry=Entry(cosm_frame, bd=2, textvariable=self.cosm3value, width=7)
        cosm3_entry.grid(row=2, column=1, ipadx=3, ipady=3)

        cosm4=Label(cosm_frame, text="Deodrants",font=("Calibri", 10))
        cosm4.grid(row=3, column=0, padx=5, pady=5)
        cosm4_entry=Entry(cosm_frame, bd=2, textvariable=self.cosm4value, width=7)
        cosm4_entry.grid(row=3, column=1, ipadx=3, ipady=3)

        cosm5=Label(cosm_frame, text="Hair Shampoo",font=("Calibri", 10))
        cosm5.grid(row=4, column=0, padx=5, pady=5)
        cosm5_entry=Entry(cosm_frame, bd=2, textvariable=self.cosm5value, width=7)
        cosm5_entry.grid(row=4, column=1, ipadx=3, ipady=3)

        #~~~~~~~~~~~~~~~~~~~~~~~~~Bill Area~~~~~~~~~~~~~~~~~~~~~
        
        bill_frame=Label(self.frame, bd=5)
        bill_frame.place(x=710, y=175, width=325, height=380)
        
        #~~~~~~~~~~~~~~~~~~~Title of the Bill Area~~~~~~~~~~~~~~~~~~~~
        
        bill_title=Label(bill_frame, text="Total Bill", font=("Calibri", 12, "bold"), bd=5)
        bill_title.pack(fill= X)
        

        #~~~~~~~~~~~~~~~Command for ScrollBar Widget~~~~~~~~~~~~~~
        
        scroll_y = Scrollbar(bill_frame,orient = VERTICAL)   #ScrollBar() widget is used for text bill area, so that it accomodate all the product at once
        self.txt = Text(bill_frame,yscrollcommand = scroll_y.set)
        scroll_y.pack(side = RIGHT,fill = Y)
        scroll_y.config(command = self.txt.yview)
        self.txt.pack(fill = BOTH,expand = 1)

        #~~~~~~~~~~~~~~~~~~~Bill Menu Frame~~~~~~~~~~~~~~~~~~~~~
        
        menu_frame=LabelFrame(self.frame, text="Bill Menu", bd=3, fg="#000000", font=("Calibri", 12, "bold"))
        menu_frame.place(x=10, y=475, relwidth=0.675, height=75)

        #~~~~~~~~~~~~~~~~~~~~~Contents of the Bill Menu Frame~~~~~~~~~~~~~~~~~~~~~~
        
        veg_total=Label(menu_frame, font=("Calibri", 12), text="Vegetable Total")
        veg_total.grid(row=0, column=0, padx=5, pady=5)
        vegtotal_entry=Entry(menu_frame, textvariable=self.totalveg, bd=2, width=10)
        vegtotal_entry.grid(row=0, column=1, ipadx=3, ipady=3)

        fru_total=Label(menu_frame, font=("Calibri", 12), text="Fruits Total")
        fru_total.grid(row=0, column=2, padx=5, pady=5)
        frutotal_entry=Entry(menu_frame, textvariable=self.totalfruit, bd=2, width=10)
        frutotal_entry.grid(row=0, column=3, ipadx=3, ipady=3)

        dry_total=Label(menu_frame, font=("Calibri", 12), text="Dairy Total")
        dry_total.grid(row=0, column=4, padx=5, pady=5)
        drytotal_entry=Entry(menu_frame, textvariable=self.totaldairy, bd=2, width=9)
        drytotal_entry.grid(row=0, column=5, ipadx=3, ipady=3)

        cos_total=Label(menu_frame, font=("Calibri", 12), text="Cosmetics Total")
        cos_total.grid(row=0, column=6, padx=5, pady=5)
        costotal_entry=Entry(menu_frame, textvariable=self.totalcosm, bd=2, width=7)
        costotal_entry.grid(row=0, column=7, ipadx=3, ipady=3)

        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Code for placing the buttons~~~~~~~~~~~~~~~~~~~~~~~ 
        
        Total_button=Button(self.frame, text="Total", fg="#000000", font=("Calibri", 12, "bold"), bd=3, command=self.total)
        Total_button.place(x=500,y=205, width=100, height=30)

        GenBill_button=Button(self.frame, text="Generate Bill", fg="#000000", font=("Calibri", 12, "bold"), bd=3, command=self.gen_bill)
        GenBill_button.place(x=600,y=205, width=100, height=30)

        Clear_button=Button(self.frame, text="Clear", fg="#000000", font=("Calibri", 12, "bold"), bd=3, command=self.clear)
        Clear_button.place(x=500,y=240, width=100, height=30)

        Exit_button=Button(self.frame, text="Exit", fg="#000000", font=("Calibri", 12, "bold"), bd=3, command=self.exit)
        Exit_button.place(x=600,y=240, width=100, height=30)

#Function to get total prices
    def total(self):
        #=================Total Food Prices
        self.totalveg_prices = ((self.veg1value.get()*20)+(self.veg2value.get()*40)+(self.veg3value.get()*20)+(self.veg4value.get()*30)+(self.veg5value.get()*20))
        self.totalveg.set("₹"+str(self.totalveg_prices))
        
        #====================Total Grocery Prices
        self.totalfruit_prices = ((self.fr1value.get()*200)+(self.fr2value.get()*150)+(self.fr3value.get()*30)+(self.fr4value.get()*100)+(self.fr5value.get()*60))
        self.totalfruit.set("₹"+str(self.totalfruit_prices))
        
        #======================Total Other Prices
        self.totaldairy_prices = ((self.d1value.get()*50)+(self.d2value.get()*60)+(self.d3value.get()*50)+(self.d4value.get()*70)+(self.d5value.get()*60))
        self.totaldairy.set("₹"+str(self.totaldairy_prices))
        
        self.totalcosm_prices = ((self.cosm1value.get()*150)+(self.cosm2value.get()*20)+(self.cosm3value.get()*325)+(self.cosm4value.get()*275)+(self.cosm5value.get()*300))
        self.totalcosm.set("₹"+str(self.totalcosm_prices))

#Function For Text Area
    def bill_info(self):  
        self.txt.delete('1.0',END)  #.delete() function will delete the content (if any) from the text area
        self.txt.insert(END,"    Welcome To Whole Super Market! \n")
        self.txt.insert(END,f"\nBill No. : {str(self.BillNo.get())}")
        self.txt.insert(END,f"\nCustomer Name : {str(self.name.get())}")
        self.txt.insert(END,"\n===================================")
        self.txt.insert(END,"\nProduct          Qty         Price")
        self.txt.insert(END,"\n===================================")

#Function to clear the bill area
    def clear(self):
        self.txt.delete('1.0',END)  #.delete() function will delete the content (if any) from the text area

#Add Product name , qty and price to bill area
    def gen_bill(self):
        self.bill_info()
        if self.veg1value.get() != 0:
            self.txt.insert(END,f"\nPotato             {self.veg1value.get()}           {self.veg1value.get()*20}")
        if self.veg2value.get() != 0:
            self.txt.insert(END,f"\nCabbage            {self.veg2value.get()}           {self.veg2value.get()*40}")
        if self.veg3value.get() != 0:
            self.txt.insert(END,f"\nTomato             {self.veg3value.get()}           {self.veg3value.get()*20}")
        if self.veg4value.get() != 0:
            self.txt.insert(END,f"\nOnions             {self.veg4value.get()}           {self.veg4value.get()*30}")
        if self.veg5value.get() != 0 :
            self.txt.insert(END,f"\nBaby Corn          {self.veg5value.get()}           {self.veg5value.get()*20}")
            
        if self.fr1value.get() != 0:
            self.txt.insert(END,f"\nApple              {self.fr1value.get()}           {self.fr1value.get()*200}")
        if self.fr2value.get() != 0:
            self.txt.insert(END,f"\nPear               {self.fr2value.get()}           {self.fr2value.get()*150}")
        if self.fr3value.get() != 0:
            self.txt.insert(END,f"\nWatermelon         {self.fr3value.get()}           {self.fr3value.get()*30}")
        if self.fr4value.get() != 0:
            self.txt.insert(END,f"\nPomegranate        {self.fr4value.get()}           {self.fr4value.get()*100}")
        if self.fr5value.get() != 0:
            self.txt.insert(END,f"\nKiwi               {self.fr5value.get()}           {self.fr5value.get()*60}")
            
        if self.d1value.get() != 0:
            self.txt.insert(END,f"\nMilk               {self.d1value.get()}           {self.d1value.get()*50}")
        if self.d2value.get() != 0:
            self.txt.insert(END,f"\nYoghurt            {self.d2value.get()}           {self.d2value.get()*60}")
        if self.d3value.get() != 0:
            self.txt.insert(END,f"\nCheese             {self.d3value.get()}           {self.d3value.get()*50}")
        if self.d4value.get() != 0:
            self.txt.insert(END,f"\nButter             {self.d4value.get()}           {self.d4value.get()*70}")
        if self.d5value.get() != 0:
            self.txt.insert(END,f"\nCustard            {self.d5value.get()}           {self.d5value.get()*60}")
            
        if self.cosm1value.get() != 0:
            self.txt.insert(END,f"\nFace Cream         {self.cosm1value.get()}           {self.cosm1value.get()*150}")
        if self.cosm2value.get() != 0:
            self.txt.insert(END,f"\nBody Soap          {self.cosm2value.get()}           {self.cosm2value.get()*20}")
        if self.cosm3value.get() != 0:
            self.txt.insert(END,f"\nBody Lotion        {self.cosm3value.get()}           {self.cosm3value.get()*325}")
        if self.cosm4value.get() != 0:
            self.txt.insert(END,f"\nDeodrants          {self.cosm4value.get()}           {self.cosm4value.get()*275}")
        if self.cosm5value.get() != 0:
            self.txt.insert(END,f"\nHair Shampoo       {self.cosm5value.get()}           {self.cosm5value.get()*300}")
        self.txt.insert(END,"\n===================================")
        self.txt.insert(END,f"\n                      Total : ₹{self.totalveg_prices+self.totalfruit_prices +self.totaldairy_prices+self.totalcosm_prices}")


    def exit(self):
        self.window.destroy()  #.destroy() function will terminate the window after the "exit" button in the GUI is clicked 



        


window=tk.Tk()  #
object=Grocery_Bill_App(window)  #this will unpack all the resources into the window of the GUI
window.mainloop()  #mainloop() function will tell Python to run the Tkinter event loop
```
